Now Let's start with new csv file

```python
import pandas as pd

df = pd.read_csv("employees.csv")
print(df)
```

```python
avg_salary = df["Salary"].mean()
print("Average Salary :", avg_salary, "\n")    

ave_salary_by_dept = df.groupby("Department")["Salary"].mean()
print(ave_salary_by_dept)
```

```python
total_salary_by_dept = df.groupby("Department")["Salary"].sum()
print(total_salary_by_dept)
```

```python
many_details = df.groupby("Department")["Salary"].agg(
    ["mean", "sum", "max", "min"]
)

print(many_details)
```
