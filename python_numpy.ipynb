```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
print(arr)

print(arr[0])
print(arr[2])
```

```python
arr = np.array([10, 20, 30, 40])

print(arr)

print(arr.ndim)  # Number of dimensions
print(arr.shape)  # Shape of the array          
print(arr.size)   # Total number of elements
print(arr.dtype)  # Data type of the elements
```

```python
arr = np.array([[10, 20, 30, 40],
                [50, 60, 70, 80]])

print(arr)
print("arr[1, 2] --> ", arr[1, 2])

print(arr.ndim)   # dimensions
print(arr.shape)  # shape
print(arr.size)   # number of elements
print(arr.dtype)  # data type
```

```python
arr_zero = np.zeros((3, 4))
print(arr_zero)
```

```python
arr_ones = np.ones((2, 3))
print(arr_ones)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(np.mean(arr))
print(np.sum(arr))
print(np.max(arr))
print(np.min(arr))
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])

new_arr = arr.reshape(2, 3)

print(new_arr)
```
